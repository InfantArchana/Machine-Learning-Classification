```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
dataset = pd.read_csv('CKD.csv')
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset, drop_first=True, dtype=int)
```

```python
dataset
```

```python
dataset["classification_yes"].value_counts()
```

```python
dataset.columns
```

```python
independent = dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
independent
```

```python
dependent = dataset['classification_yes']
dependent
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(independent, dependent, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

param_grid = {'penalty':['l2'],
              'solver':['lbfgs', 'liblinear', 'newton-cg', 'saga'],
              'C':[0.1, 10, 100, 1000],
              'max_iter':[5000, 6000]}

grid = GridSearchCV(LogisticRegression(), param_grid, refit = True, verbose = 3, n_jobs=-1, scoring='f1_weighted')

grid.fit(x_train, y_train)
```

```python
re=grid.cv_results_
```

```python
table=pd.DataFrame.from_dict(re)
table
```

```python
grid_predictions = grid.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, grid_predictions)
```

```python
print("The confusion Matrix:\n",cm)
```

```python
from sklearn.metrics import classification_report
clf_report = classification_report(y_test, grid_predictions)
```

```python
print("The Classification Report:\n", clf_report)
```

```python
from sklearn.metrics import f1_score
f1_macro=f1_score(y_test, grid_predictions, average='weighted')
```

```python
print("The f1_macro value for best parameter {}:".format(grid.best_params_),f1_macro)
```

```python
from sklearn.metrics import roc_auc_score

roc = roc_auc_score(y_test, grid.predict_proba(x_test)[:,1])
```

```python
print("The ROC curve from prediction scores:", roc)
```
